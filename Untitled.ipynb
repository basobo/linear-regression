```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
df=pd.read_csv('datasets_85008_196530_Ecommerce Customers.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
data=df[['Time on Website','Yearly Amount Spent']]
```

```python
data.head(5)
```

```python
def fit_and_plot_model(data, model):
    features = data.drop('Time on Website', axis=1)
    target = data['Yearly Amount Spent']
    model.fit(features, target)
    plt.scatter(data['Time on Website'], data['Yearly Amount Spent'])
    plt.plot(data['Time on Website'], model.predict(features), 'k--')
    return model
```

```python
# Unconstrained Decision Tree
model = fit_and_plot_model(data, DecisionTreeRegressor())
```

```python
# Original values of y
actual_y = df['Yearly Amount Spent']

# Predicted values of y
predicted_y = model.predict(data.drop('Yearly Amount Spent', axis=1))

# Total difference between actual and predicted values of y
np.sum( np.absolute( actual_y - predicted_y ) )
```

```python
# Exports the tree structure
from sklearn.tree import export_graphviz

# Allows us to render the diagram in Jupyter notebook
from graphviz import Source
```

```python
from IPython.display import Image

def plot_and_save_decision_tree(model, out_filename):
    # Export tree structure and save to Source object
    graph = Source( export_graphviz(model, out_file=None) )

    # Pipe to PNG bytes
    tree_png = graph.pipe(format='png')

    # Write to PNG file
    with open(out_filename,'wb') as f:
        f.write(tree_png)

    # Display image
    return Image(tree_png)
```

```python
plot_and_save_decision_tree(model, 'tree_graph.png')
```
